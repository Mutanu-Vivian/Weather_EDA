```python
df['Weather']=df['Weather'].replace(to_replace='.*Drizzle.*',value='Drizzle',regex=True)
df['Weather']=df['Weather'].replace(to_replace='.*Cloudy.*',value='Cloudy',regex=True)
df['Weather']=df['Weather'].replace(to_replace='.*Rain.*',value='Rain',regex=True)
df['Weather']=df['Weather'].replace(to_replace='.*Clear.*',value='Clear',regex=True)
df['Weather']=df['Weather'].replace(to_replace='.*Snow.*',value='Snow',regex=True)
df['Weather']=df['Weather'].replace(to_replace='.*Fog.*',value='Fog',regex=True)
df['Weather']=df['Weather'].replace(to_replace='.*Haze.*',value='Haze',regex=True)
df['Weather']=df['Weather'].replace(to_replace='.*Thunderstorms.*',value='Thunderstorms',regex=True)
```

**_Based on the nature of the Weather descriptions in the dataset, that had 2 descriptions for the same record such as 'Freezing drizzle' or 'Mostly Cloudy', I decided to write the code above to replace the inconsistent weather descriptions, with defined names for better analysis._**

```python
sns.boxplot(x="Weather",y="Temp_C",data=df)
```

**_The box plot above diplays the distribution of different weather conditions. It can be seen that the distribution of temperature across all weather categories are distict. Outlier temperatures can be spotted in 'Snow' and 'Haze' weather categories.  
The anomalies observed in the box plot might be a result of standardising the terms in the code run to replace ambiguous weather descriptions with defined terms._**

```python
weather_counts=df["Weather"].value_counts()

sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
plt.pie(weather_counts, labels=weather_counts.index,autopct='%1.2f%%', startangle=140)
plt.title('Weater Conditions Distribution')
plt.axis('equal')
plt.show()
```

```python
sns.histplot(df['Temp_C'],kde=True)
plt.title('Temperature Distribution')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(data=df, x='Press_kPa', y='Wind Speed_km/h')
plt.title('Air Pressure vs Wind Speed')
plt.xlabel('Air Pressure')
plt.ylabel('Wind Speed')
plt.show()
```

```python
plt.figure(figsize=(14,7))
df['Temp_C'].plot(label='Temperature')
df['Dew Point Temp_C'].plot(label='Dew Point')
plt.title('Time series of Temperature and Dew Point')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend() 
plt.show()
```

**_There is a clear pattern of weather observed when this time series is generated to compare the variables Temperature and Dew Point. Three distinct seasons can observed in the Temperature and Humidity observations as follows:  
   Low levels- January to April  
   Rising levels- May to September  
   Decreasing levels-October to December.  
The pattern then repeats itself in January, as per the observation._**

```python
selected_columns=df[['Temp_C','Dew Point Temp_C','Rel Hum_%','Wind Speed_km/h','Visibility_km','Press_kPa']]
correlation_matrix=selected_columns.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidth=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

**_This is a correlation heatmap that compares correlation between the weather variables that have datatypes as integers. These are Temperature, Dew Point, Humidity, Windspeed,Visibility and Atmospheric Pressure.The negative correlation coefficients signify a negative correlation between the variables, with those close to -1 having a strong negative correlation, and those closer to 0 show a weak negative correlation. Consequently,the positive coefficients signify a strong positive correlation for the coefficients close to +1, and a weak positive correlation for coefficients close to 0._**
